# Setup
First, let's import TensorFlow and other necessary libraries. Run the cell below to import them.


```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

## Dataset Preparation
We'll use a synthetic dataset for binary classification. The 'make_moons' function from scikit-learn generates a two-moon dataset, which is ideal for this experiment.

```python
# Generate synthetic data
X, y = make_moons(n_samples=1000, noise=0.2, random_state=42)

# Splitting the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Visualize the dataset
# selects first two columns of X_train to plot
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.RdBu)
plt.title("Training Data")
plt.show()
```

## Model Building
Now, let's define a function to create a simple neural network model. You will modify this function later to experiment with different activation and loss functions.

```python
import tensorflow as tf

def build_model(activation_function='relu', loss_function='binary_crossentropy'):
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(10, activation=activation_function, input_shape=(2,)),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss=loss_function, metrics=['accuracy'])
    tf.keras.utils.plot_model(model, to_file='model.png', show_shapes=True)
    return model

model = build_model()
model.summary()
```

# Training and evaluating the model
Let's train the model and see how well it solves the problem.

```python
# Function to train, evaluate, and visualize model predictions along with ground truth
def train_and_evaluate_model(model, epochs=100):
    history = model.fit(X_train, y_train, epochs=epochs, verbose=0, validation_split=0.2)
    
    # Evaluating the model on the test set
    test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
    
    # Plotting training history
    plt.figure(figsize=(18, 5))

    plt.subplot(1, 3, 1)
    plt.plot(history.history['accuracy'], label='Accuracy (training data)')
    plt.plot(history.history['val_accuracy'], label='Accuracy (validation data)')
    plt.title('Training and Validation Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()

    # Predicting on test data
    predictions = model.predict(X_test) > 0.5

    # Plotting model predictions
    plt.subplot(1, 3, 2)
    plt.scatter(X_test[:, 0], X_test[:, 1], c=predictions[:, 0], cmap=plt.cm.coolwarm)
    plt.title('Model Predictions on Test Data')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

    # Plotting ground truth
    plt.subplot(1, 3, 3)
    plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=plt.cm.coolwarm)
    plt.title('Ground Truth for Test Data')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

    plt.show()

    print(f"Test Loss: {test_loss:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f}")

# Example of using the function with a specific activation and loss function
model = build_model(activation_function='relu', loss_function='binary_crossentropy')
train_and_evaluate_model(model, epochs=100)

# MSE: 0.0814 / 0.8900
# binary_crossentropy: 0.2408 / 0.8967
# vi vill minimera loss och maximera accuracy
```

# Notes & Thoughts





- **High Test Loss**: Indicates that the model's predictions are far off from the actual values. A higher loss signifies worse performance of the model on the test data.
    
- **Low Test Accuracy**: Shows that the model is not classifying the inputs correctly. In the context of binary classification, low accuracy means the model is often incorrect in distinguishing between the two classes.
    
- **Difference Between Test Loss and Test Accuracy**:
    
    - Loss is a measure of how wrong the model's predictions are (quantitatively), often represented by a continuous function where lower values are better.
    - Accuracy is a measure of the model's performance in terms of correct classifications, usually expressed as a percentage where higher is better.
    - In some cases, a model might have a relatively low loss but still not achieve high accuracy. This could happen if many predictions are close but not correct enough to be classified accurately.

Ideally, you want to minimize loss and maximize accuracy, but the relationship between them isn't always straightforward. Sometimes improving accuracy might not significantly lower the loss and vice versa, depending on the complexity of the model, the nature of the data, and the specific loss function used.
